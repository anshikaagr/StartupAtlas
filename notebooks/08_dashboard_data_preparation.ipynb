# Dashboard Data Preparation

## Objective

Prepare clean summary datasets for the StartupAtlas dashboard.

These datasets will be used for interactive visualizations and startup ecosystem exploration.

```python
import pandas as pd
import numpy as np

df = pd.read_csv(
    "../data/raw/investments_VC.csv",
    encoding="latin1"
)

df.columns = df.columns.str.strip()

df["funding_total_usd"] = (
    df["funding_total_usd"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("-", "", regex=False)
    .str.strip()
)

df["funding_total_usd"] = pd.to_numeric(
    df["funding_total_usd"],
    errors="coerce"
)

df.shape
```

```python
df.to_csv(
    "../data/processed/startup_dna_dataset.csv",
    index=False
)
```

## Dashboard Dataset: Country Summary

Create country-level startup metrics for dashboard visualizations.

```python
country_summary = (
    df.groupby("country_code")
    .agg(
        startups=("name", "count"),
        total_funding=("funding_total_usd", "sum"),
        avg_funding=("funding_total_usd", "mean")
    )
    .reset_index()
)

country_summary.head()
```

```python
country_summary.to_csv(
    "../data/processed/country_summary.csv",
    index=False
)
```

## Dashboard Dataset: Market Summary

Create industry-level startup metrics for dashboard analysis.

```python
market_summary = (
    df.groupby("market")
    .agg(
        startups=("name", "count"),
        total_funding=("funding_total_usd", "sum"),
        avg_funding=("funding_total_usd", "mean")
    )
    .reset_index()
)

market_summary.head()
```

```python
market_summary.to_csv(
    "../data/processed/market_summary.csv",
    index=False
)
```

## Dashboard Dataset: Startup Status Summary

Create startup outcome statistics for dashboard charts.

```python
status_summary = (
    df["status"]
    .value_counts()
    .reset_index()
)

status_summary.columns = [
    "status",
    "count"
]

status_summary
```

```python
status_summary.to_csv(
    "../data/processed/status_summary.csv",
    index=False
)
```

## Dashboard Dataset: Funding Trend Summary

Create yearly funding statistics for dashboard trend analysis.

```python
funding_summary = (
    df.groupby("founded_year")
    .agg(
        total_funding=("funding_total_usd", "sum")
    )
    .reset_index()
)

funding_summary.head()
```

```python
funding_summary.to_csv(
    "../data/processed/funding_summary.csv",
    index=False
)
```
