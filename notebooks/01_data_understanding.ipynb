```python
top_funding = (
    df.groupby('market')['funding_total_usd']
      .sum()
      .sort_values(ascending=False)
      .head(20)
)

top_funding
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

top_funding.plot(kind='bar')

plt.title("Top Startup Markets by Total Funding")
plt.xlabel("Market")
plt.ylabel("Funding (USD)")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.bar(
    x=top_funding.index,
    y=top_funding.values,
    title="Top Startup Markets by Total Funding"
)

fig.show()
```
