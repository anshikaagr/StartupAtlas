```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "../data/raw/investments_VC.csv",
    encoding="latin1"
)

df.columns = df.columns.str.strip()
```

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
df.columns.tolist()
```

```python
df['status'].value_counts()
```

```python
df['funding_rounds'].describe()
```

```python
df['market'].value_counts().head(20)
```

```python
df['funding_total_usd'] = (
    df['funding_total_usd']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('-', '', regex=False)
    .str.strip()
)

df['funding_total_usd'] = pd.to_numeric(
    df['funding_total_usd'],
    errors='coerce'
)
```

```python
df['funding_total_usd'].describe()
```

```python
top_funding = (
    df.groupby('market')['funding_total_usd']
      .sum()
      .sort_values(ascending=False)
      .head(20)
)

top_funding
```

```python
fig = px.bar(
    x=top_funding.index,
    y=top_funding.values,
    title="Top Startup Markets by Total Funding",
    labels={
        "x": "Market",
        "y": "Total Funding (USD)"
    }
)

fig.show()
```
